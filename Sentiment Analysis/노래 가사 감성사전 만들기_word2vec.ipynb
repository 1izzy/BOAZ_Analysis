```python
#모델 학습
model = Word2Vec(data['가사전처리'], window=5, min_count=3, size=300)
```

```python
model.wv.similar_by_word('슬픔', topn = 10)
```

```python
#0.75 유사도인 단어 집합 확인
emotion = ['슬픔', '행복', '짜증']
emotion_list = []
for idx in range(len(emotion)):
  emotion_one = []
  for word in model.wv.similar_by_word(emotion[idx], topn = 300):
    if word[1] >= 0.75:
      emotion_one.append(word[0])
  emotion_list.append(emotion_one)
```

```python
whole_emotion = []
for j in range(data.shape[0]):
  data_dic = {'sad': 0, 'happy': 0, 'angry':0}
  for k in ['sad', 'angry','happy']:
    for i in emotion_dic[k]:
      if i == data['가사전처리'][j][0]:
        data_dic[k] += 1
        
  emotions = [data_dic['sad'], data_dic['happy'], data_dic['angry']]
  whole_emotion.append(emotions)
  print(emotions)
```

```python
#1있는 리스트
count=0
for i in whole_emotion :
  for j in range(2) :
    if i[j] == 1 :
      count +=1
print(count)
```

```python
# matrix에 1 들어가있는 데이터 프레임 뽑기 
data['matrix']=whole_emotion

df=pd.DataFrame()
col = 0
for i in whole_emotion :
  for j in range(3) :
    if i[j] == 1 :
      df=df.append(data.loc[col])
  col +=1
print(df)
```
